```python
df = pd.read_csv("datafest2018-Updated-April12.csv")

temp = df.loc[:,['jobId','city']].drop_duplicates()['city'].value_counts()
cityset = set(temp.loc[temp>1000].index.tolist())

temp = df.loc[:,['jobId','industry']].drop_duplicates()['industry'].value_counts()
industryset = set(temp.loc[temp>1000].index.tolist())

df = df.loc[df['country']=='US']
df = df.loc[df['city'].isin(cityset)]
df = df.loc[df['industry'].isin(industryset)]

df = df.loc[:,['date', 'jobId', 'country', 'stateProvince', 'city','avgOverallRating', 'industry', 'experienceRequired', 'supervisingJob','educationRequirements', 'jobAgeDays', 'clicks']]

df.head()
```
